## Описание проекта

Интернет-магазине «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы. Задача - выявить определяющие успешность игр закономерности и разработать рекомендации по развитию наиболее популярного продукта для планирования рекламных кампаний.

 #### Описание данных
 * Name — название игры
 * Platform — платформа
 * Year_of_Release — год выпуска
 * Genre — жанр игры
 * NA_sales — продажи в Северной Америке (миллионы проданных копий)
 * EU_sales — продажи в Европе (миллионы проданных копий)
 * JP_sales — продажи в Японии (миллионы проданных копий)
 * Other_sales — продажи в других странах (миллионы проданных копий)
 * Critic_Score — оценка критиков (максимум 100)
 * User_Score — оценка пользователей (максимум 10)
 * Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Заменим названия столбцов (приведем их к нижнему регистру).

```python
df.columns = df.columns.str.lower()
df.info()
```

В столбце `genre` поменяем заглавные на строчные.

```python
df['genre'].str.lower().unique()
```

```python
df['year_of_release'] = df['year_of_release'].astype('Int32') #заменим тип данных на Int
df.info()
```

В столбце `user_score` есть значение-аббревиатура 'tbd'. Возможно, это t.b.d.= to be defined - подлежит определению. Заменим 'tbd' на NaN, как на не определенные значения в наших данных.

```python
df[df['user_score'] == 'tbd'].count()
```

2424 строк со значением 'tbd' в столбце. Изменим тип данных в столбце на числовой. Параметр errors='coerce' принудительно заменит на NaN некорректные значения 'tbd'.

```python
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce') # `tbd` переводим в NaN
df['user_score'] = df['user_score'].astype('Float32')
df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].astype('Float32')
df.info()
```

Посмотрим пропущенные значения.

```python
df.isna().sum()
```

В столбце `name` 2 пропущенных значения. Удалим пропущенные значения.

```python
df.dropna(subset = ['name'], inplace = True)
df.info()
```

Посмотрим: есть ли дубликаты в столбцах `name`, `platform` и `year_of_release`.

```python
df[['name', 'platform', 'year_of_release']].duplicated().sum()
```

Посмотрим на дубликаты и удалим.

```python
df_dupl = df[df[['name', 'platform', 'year_of_release']].duplicated()]
print(format(df_dupl))
```

Удалим строку 16230 из нашей таблицы.

```python
df.drop(labels = [16230], axis = 0, inplace = True)
df.info()
```

```python
df['rating'] = df['rating'].fillna('unknown')
df.isna().sum()
```

```python
df.head()
```

Посчитаем суммарные продажи во всех регионах и запишем их в столбец `all_sales`.

```python
df['all_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
df.sample()
```

Шкалы оценок отзывов пользователей и критиков разные: оценки пользователей находятся в диапазоне от 0 до 100 баллов, критиков - от 0 до 10. Приведем их к сравнимым оценочным системам, умножив значения в столбце `user_score` на 10, и поместим данные в новый столбец `user_score_new`.

```python
df['user_score_new'] = df['user_score'] * 10
```

### Изменения продаж по платформам
Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам. Ответим на вопрос: за какой характерный срок появляются новые и исчезают старые платформы.

```python
platforms_sales = df.groupby('platform')['all_sales'].count().sort_values(ascending = False)
# platforms_sales.sort_values(ascending = False)
platforms_sales
```

Возьмем список самых топовых платформ (первая 10-ка) с суммарными продажами большими чем 820 млн проданных копий и создадим сводную таблицу `top_platforms` - распределение по годам.

```python
top_platforms = df[df['platform'].isin(platforms_sales.index[:10])].pivot_table(index='year_of_release', columns='platform', values='name', aggfunc='count')
top_platforms
```

```python
top_platforms.plot(grid=True, figsize=(16,8), xlim=(1993,2016))
plt.legend(title ='название платформы')
plt.title('Распределение топовых платформ по годам')
plt.xlabel('годы')
plt.ylabel('количество, шт')
plt.show()
```

Платформы доминируют в своей популярности в течение некоторого периода времени, сменяя друг друга или появляясь одновременно на рынке. Исходя из табличных данных и графиков, средний срок жизнеспособности игровой платформы составляет около 10 лет. (Для платформы PC данные за 1994-2000 гг можно отбросить, так как выпуск игр для этой платформы находился в районе нулевых значений).

Поскольку игровая индустрия развивается быстрыми темпами, для дальнейшего прогнозирования на 2017 год актуальным периодом будем считать диапазон в 3 года - период, в который происходит рост платформ, то есть период с 2013 по 2016 гг.

Создадим таблицу `df_prediction` с данными за эти годы.

```python
df_prediction = df[df['year_of_release'] >= 2013]
df_prediction.info()
```

Посмотрим, какие платформы лидируют по продажам.

### Влияние отзывов пользователей и критиков на продажи платформы `X360`

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

Возьмем одну из самых прибыльных платформ `X360`.

```python
df_x360 = df_prediction_vyborka[df_prediction_vyborka['platform'] == 'X360']
df_x360.plot(x='critic_score', y='all_sales', kind='scatter', color='red')
plt.xlabel('оценка критиков, баллы', fontsize=14)
plt.ylabel('продажи, млн', fontsize=14)
plt.title('Зависимость оценки критиков от продаж\nна платформе X360', fontsize=16)

df_x360.plot(x='user_score_new', y='all_sales', kind='scatter')
plt.xlabel('оценка пользователей, баллы', fontsize=14)
plt.ylabel('продажи, млн', fontsize=14)
plt.title('Зависимость оценки пользователей от продаж\nна платформе X360', fontsize=16)
plt.show()
```

```python
df_x360[['critic_score', 'all_sales']].corr()
```

```python
df_x360[['user_score_new', 'all_sales']].corr()
```

Коэффициенты корреляции между продажами и отзывами критиков и пользователей меньше 0.5, выводы о какой-либо явной взаимосвязи мы сделать не можем, хотя по диаграмме рассеивания видны некоторые закономерности.
Исключив выбросы, получим следующие коэффициенты корреляции.

```python
df_x360_cut = df_x360[df_x360['all_sales'] < 1.75]
df_x360_cut[['critic_score', 'all_sales']].corr()
```

```python
df_x360_cut[['user_score_new', 'all_sales']].corr()
```

Коэффициенты выросли после очистки выбросов, но очевидной взаимосвязи между продажами и отзывами также не наблюдается (хотя у оценок критиков и продаж есть умеренная корреляция 0.47).

## Проверка гипотез

Проверим следующие гипотезы:

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (экшен-игры) и Sports (спортивные соревнования) разные.


Задайте самостоятельно пороговое значение alpha.
Поясните:
•	Как вы сформулировали нулевую и альтернативную гипотезы;
•	Какой критерий применили для проверки гипотез и почему.

### Гипотеза 1
Сформулируем нулевую гипотезу о том, что средние пользовательские рейтинги платформ `XOne` и `PC` равны, и альтернативную
гипотезу, что средние пользовательские рейтинги платформ `XOne` и `PC` различаются. Пороговое значение alpha зададим в 0.05.

Чтобы проверить гипотезу о равенстве среднего наших генеральных совокупностей по взятым из них выборкам, применим метод
scipy.stats.ttest_ind(), так как размеры выборок достаточно большие. В нашем случае размеры выборки (182 для `XOne` и 155 для `PC`) и дисперсии (стандартные отклонения: 1.38 и 1.74, соответственно) не равны в наших группах, поэтому передадим параметр: equal_var = False.

Сделаем срезы из таблицы `df_prediction` с учетом платформ и исключим нулевые значения в `user_score`

```python
rating_xone = df_prediction.query('platform == "XOne" & user_score != 0')
rating_xone.info()
```

```python
rating_xone.describe()
```

```python
rating_pc = df_prediction.query('platform == "PC" & user_score != 0')
rating_pc.info()
```

```python
rating_pc.describe()
```

```python
df_prediction
```

```python
temp1, temp2 = df_prediction.query('platform == "XOne"'), df_prediction.query('platform == "PC"')
temp1[temp1.user_score == 0].user_score.count(), temp2[temp2.user_score == 0].user_score.count()
```

```python
alpha = .05
results = st.ttest_ind(rating_pc['user_score'], rating_xone['user_score'], equal_var = False)
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print ('Отвергаем нулевую гипотезу')
else:
    print ('Не получилось отвергнуть нулевую гипотезу')
```

Значение pvalue в 0.15 дает достаточное основание для выбранного уровня значимости принять нулевую гипотезу. Значит, можно сделать вывод, что средние пользовательские рейтинги платформы `XOne` не отличаются от средних пользовательских рейтингов платформы `PC`.

### Гипотеза 2

Сформулируем нулевую гипотезу о том, что средние пользовательские рейтинги жанров `Action` и `Sports` равны, и альтернативную
гипотезу, что средние пользовательские рейтинги жанров `Action` и `Sports` различаются. 

Для проверки гипотезы, применим метод scipy.stats.ttest_ind() . Методу, как и в проверке первой гипотезы, передадим параметр по
умолчанию: equal_var = False (размеры выборки 389 для `Action` и 160 для `Sports` и стандартные отклонения: 1.33 и 1.78, соответственно) не равны в наших группах. Пороговое значение alpha зададим также в 0.05.

```python
rating_action = df_prediction.query('genre == "Action" & user_score != 0')
rating_action.info()
```

```python
rating_action.describe()
```

```python
rating_sports = df_prediction.query('genre == "Sports" & user_score != 0')
rating_sports.info()
```

```python
rating_sports.describe()
```

```python
alpha = .05
results = st.ttest_ind(rating_action['user_score'], rating_sports['user_score'], equal_var = False)
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print ('Отвергаем нулевую гипотезу')
else:
    print ('Не получилось отвергнуть нулевую гипотезу')
```

Значение pvalue в 1.45 дает достаточное основание для выбранного уровня значимости (даже при пороговом значении pvalue в 0.1) отвергнуть нулевую гипотезу.

Также популярен жанр `Action`- 29% от продаж в Японии. Вкусовые предпочтения по жанрам также отличаются у японцев от Европы и Америки, жанр `Fighting`, к примеру, не вошел в топ-5 мира.

- Рейтинг ESRB влияет на европейских и американских игроков, в отличие от Японии, в которой суммарные продажи у игр без рейтинга выше.

- Корреляционной очевидной зависимости между отзывами пользователей, критиков и суммарными продажами не наблюдается.

- Средние пользовательские рейтинги жанров `Action` и `Sports` различаются.
- Средние пользовательские рейтинги платформы `XOne` не отличаются от средних пользовательских рейтингов платформы `PC`.
