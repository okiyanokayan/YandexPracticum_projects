Посмотрим медиану распределения столбца `ceiling_height` и заполним пропуски значением в 2.65. Выбор медианы обоснован тем, что в этом столбце мы имеем выбивающиеся значения (100 м, к примеру), и среднее завышено. А разбивать высоту потолков по категориям не имеет смысла, поскольку мы не знаем более полной информации об объектах недвижимости (серию дома, типовые харахтеристики и т.п.)

Значения в столбце большие 10 м (представляемые значения высоты потолка двух-трехуровневых элитных объектов недвижимости), уменьшим на порядок, а значения меньшие 2.20 м (критерий минимальной высоты потолка типового жилья) , заменим также на медианные значения, предполагая, что это ошибки ввода.

```python
df.sort_values(by = 'ceiling_height', ascending = False)
```

Сделаем замену пропущенных значений медианой.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(2.65)
```

Напишем функцию для замены значений в столбце `ceiling_height`

```python
def ceiling_height_change(height):
    """
    Возвращает высоту потолка по значению height, используя правила:
    - 2.65, если height < 2.20;
    - height/10, если height >10.
    """
    if height > 10:
        return height/10
    if 2.20 <= height <= 10:
        return height
    return 2.65

df['ceiling_height'] = df['ceiling_height'].apply(ceiling_height_change)
df['ceiling_height'].describe()
```

Обратимся к столбцу `living_area`, в котором 8% пропущенных значений от общего кол-ва. Жилая площадь(`living_area`) зависит от общей площади жилого помещения(`total_area`). Поэтому при заполнении пропущенных значений `living_area` используем метод заполнения пропусков  по медианному значению с учетом общей площади.

```python
df['living_area'] = df['living_area'].fillna(df.groupby('total_area')['living_area'].transform('median'))
```

Заполним пропущенные значения в столбце `kitchen_area` путем нахождения разницы между общей и жилой площадью. Очевидно, что значения будут завышенными, потому что в наших рассчетах не будут учитываться площади остальных нежилых помещений. Также будем считать, что аппартаменты имеют кухню, как выделенный метраж.

```python
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area']-df['living_area'])
```

## Расчёты и добавление результатов в таблицу

Рассчитаем цену квадратного метра и занесем данные в столбец `price_metr_sq`.

```python
df['price_metr_sq'] = (df['last_price'] / df['total_area']).round(1)
```

а данные отношения площади кухни к общей в столбец `ratio_kitchen_total_area`.

(ссылка тут: [Минимальные площади помещений квартиры](https://gsps.ru/poleznoe/minimalnaya-ploshchad-pomeshcheniy-kvartiry.php#:~:text=%E2%96%BA%20%D0%A1%D0%BF%D0%B0%D0%BB%D1%8C%D0%BD%D0%B8%20%E2%80%94%208%20%D0%BC%C2%B2%20) )

Добавим столбец `floor_category` - этаж квартиры по категориям:
первый, если `floor` = 1;
последний, если `floor` = `floors_total` и `floors_total` != 1;
другой - во всех остальных случаях.

86 пропусков в колонке `floor_category` оставим без изменения, хотя можно было бы присвоить им значения `floor`.

```python
df['floor_category'] = 0

floor_conditions = [
    (df['floor'] == 1) & (df['floors_total'] != 1),
    (df['floor'] == df['floors_total']),
    (df['floor'] != 1) & (df['floor'] != df['floors_total'])
]
floor_categories = ['первый', 'последний', 'другой']

df['floor_category'] = np.select(floor_conditions, floor_categories)
df[df['floor_category'].isna()]
```

Добавим столбец `floor_category_numeric`, присвоим категории этажности числовые коды-значения:

'последний' - 3.

```python
df['floor_category_numeric'] = 0
df.loc[df['floor_category'] == 'первый', 'floor_category_numeric'] = 1
df.loc[df['floor_category'] == 'последний', 'floor_category_numeric'] = 3
df.loc[df['floor_category'] == 'другой', 'floor_category_numeric'] = 2
```
